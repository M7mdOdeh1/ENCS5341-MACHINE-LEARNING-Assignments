## Part 1: Read the dataset and examine how many features and examples does it have?

In this section, I load the `cars.csv` dataset using Pandas and examine its structure by reading the CSV file into a Pandas DataFrame and then using the `shape` attribute to extract the dimensions of the DataFrame:

- The number of examples (rows) in the dataset.
- The number of features (columns) present.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'cars.csv'
cars_df = pd.read_csv(file_path)

# Determine the number of columns and rows
num_features = cars_df.shape[1]
num_examples = cars_df.shape[0]
print("Number of features: " , num_features)
print("Number of examples: " , num_examples)
```

## Part 2: Are there features with missing values? How many missing values are there in each one?

We performed a check for missing values in the dataset to identify any features with incomplete data. The analysis revealed missing values in the following features:
- `horsepower`: Contains 6 missing values.
- `origin`: Contains 2 missing values.

```python
# Checking for missing values in each feature
missing_values = cars_df.isnull().sum()
#check if there any missing values
if missing_values.empty: 
    print("There is no missing values")
else: 
    print(missing_values.to_string())
```

## Part 3: Fill the missing values in each feature using a proper imputation method (for example: fill with mean, median, or mode)

To address the missing values in the dataset, we employed appropriate imputation methods for each feature:

- **Horsepower**: Since this is a numerical feature, we used the median value for imputation, because the median is less affected by outlier than the mean.
  
- **Origin**: Being a categorical feature, the mode (the most frequent value) was used for imputation.


```python
# Choosing imputation methods for each feature with missing values
# For 'horsepower', we'll use the median (since it's a numerical feature and median is less affected by outliers)
# For 'origin', we'll use the mode (since it's likely a categorical feature)

horsepower_median = cars_df['horsepower'].median()
origin_mode = cars_df['origin'].mode()[0]

# Fill missing values
cars_df['horsepower'].fillna(horsepower_median, inplace=True)
cars_df['origin'].fillna(origin_mode, inplace=True)

# Check if all missing values are filled
remaining_missing_values = cars_df.isnull().sum()
print(remaining_missing_values.to_string())
```

## Part 6: Support your answer for part 5 by using a quantitative measure.


```python
# Function to calculate Pearson's first coefficient of skewness
def pearsons_skewness(series):
    mean = series.mean()
    median = series.median()
    std = series.std()
    skewness = (3 * (mean - median)) / std
    return skewness

# Calculate the skewness for each feature
skewness_acceleration = pearsons_skewness(cars_df['acceleration'])
skewness_horsepower = pearsons_skewness(cars_df['horsepower'])
skewness_mpg = pearsons_skewness(cars_df['mpg'])

# Print the results
print(f"Pearson's skewness for acceleration: {skewness_acceleration}")
print(f"Pearson's skewness for horsepower: {skewness_horsepower}")
print(f"Pearson's skewness for mpg: {skewness_mpg}")
```

To complement our visual analysis, we applied Pearson's second coefficient of skewness to provide a quantitative measure of the skewness for the 'acceleration', 'horsepower', and 'mpg' distributions. This coefficient is a measure of the asymmetry of a distribution, where a value closer to zero indicates a distribution that is more symmetrical and thus more similar to a Gaussian distribution. As we can see, skewness for acceleration is too close to zero.

## part 8: Implement the closed form solution of linear regression and use it to learn a linear model to predict the ‘mpg’ from the ‘horsepower’. Plot the learned line on the same scatter plot you got in part 7.

```python
# Extract the 'horsepower' column as a NumPy array
horsepower = cars_df['horsepower'].values
mpg = cars_df['mpg'].values

# Now you can add the column of ones for the intercept term
X = np.c_[np.ones(horsepower.shape[0]), horsepower]

# Calculating the coefficients of the linear regression model
w = np.linalg.inv(X.T @ X) @ X.T @ mpg
print(w)
# Scatter plot of horsepower vs mpg
plt.figure(figsize=(10, 6))
sns.scatterplot(x=horsepower, y=mpg)

# Regression line
x_values = np.linspace(min(horsepower), max(horsepower), 100)
y_values = w[0] + w[1] * x_values
plt.plot(x_values, y_values, color='red')

plt.title('Horsepower vs MPG with Regression Line')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.show()
```

In the code, I implemented a simple linear regression model to predict 'MPG' rom 'horsepower'. Here's a breakdown of what was done:

1. **Data Preparation**: To account for the intercept term in our linear model, we added a column of ones to `X`.

2. **Model Coefficients**: Using the normal equation method, we calculated the  parameters of our linear regression model. The normal equation is a closed-form solution that involves inverting the matrix product of the transpose of `X` and `X` itself, and then multiplying it by the transpose of `X` and the vector of target values `y`.

3. **Plotting**: After computing the model parameters, we plotted the original data as a scatter plot to visualize the relationship between 'horsepower' and 'MPG'. We then plotted our regression line to represent the model we learned. This line was plotted using the calculated coefficients to provide a visual representation of our predictive model.

4. **Analysis**: The resulting plot showed a negative correlation between 'horsepower' and 'MPG' as indicated by the downward slope of the regression line. This suggests that, generally, cars with higher horsepower have lower MPG, consistent with our understanding of the trade-off between power and fuel efficiency.


## Part 9: Repeat part 8 but now learn a quadratic function of the form : f = w0 + w1x + w2x^2.

```python
# Prepare the feature matrix for quadratic regression
X_quad = np.column_stack((np.ones(len(cars_df)), cars_df['horsepower'], cars_df['horsepower']**2))
y = cars_df['mpg'].values

# Compute the coefficients for the quadratic regression model
w_quad = np.linalg.inv(X_quad.T @ X_quad) @ X_quad.T @ y
print(w_quad)
# Scatter plot of horsepower vs mpg
plt.figure(figsize=(10, 6))
sns.scatterplot(x='horsepower', y='mpg', data=cars_df)

# Plot the quadratic regression curve
x_values = np.linspace(min(cars_df['horsepower']), max(cars_df['horsepower']), 100)
y_values_quad = w_quad[0] + w_quad[1] * x_values + w_quad[2] * x_values**2
plt.plot(x_values, y_values_quad, color='green')

plt.title('Horsepower vs MPG with Quadratic Regression Curve')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.show()
```

In the quadratic regression approach, we first transformed our input feature 'horsepower' to include both the original feature and its square. This was accomplished by creating a new feature matrix `X_quad`:

1. We started with our original input feature 'horsepower'.
2. We then squared the 'horsepower' values to capture the quadratic term.
3. These two sets of values, along with a column of ones for the intercept term, were combined to form the feature matrix `X_quad`.

With `X_quad` representing our transformed features, we applied linear regression to this new feature space. Despite the model being quadratic in nature, the regression process remains linear with respect to the coefficients. 


## Part 10: Repeat part 8 (simple linear regression case) but now by implementing the gradient descent algorithm instead of the closed form solution.


```python
def gradient_descent(X, y, alpha=0.0001, iterations=1000000):
    m = len(y)  # Number of training examples
    w = np.zeros(2)  # Initialize coefficients
    
    for _ in range(iterations):
        predictions = X.dot(w)
        errors = predictions - y
        gradient = X.T.dot(errors) / m
        w -= alpha * gradient  # Update coefficients
    
    return w

# Preparing the data
X = np.column_stack((np.ones(len(cars_df)), cars_df['horsepower']))
y = cars_df['mpg'].values

# Running gradient descent
w_gd = gradient_descent(X, y)

# Print the coefficients from gradient descent
print(w_gd)

# Plot the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X[:, 1], y=y)

# Generate a range of horsepower values for plotting the regression line
x_values = np.linspace(min(X[:, 1]), max(X[:, 1]), 100)
y_values = w_gd[0] + w_gd[1] * x_values

# Plot the regression line
plt.plot(x_values, y_values, color='red')

plt.title('Horsepower vs MPG with Gradient Descent Linear Regression')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.show()
```

We applied gradient descent to learn the parameters of a linear regression model predicting 'MPG' from 'horsepower'. The algorithm iteratively adjusted the weights to minimize the error between predicted and actual 'MPG' values over 1,000,000 iterations, with a learning rate of 0.0001.

The resulting plot shows the regression line against the scatter plot of data points. The negative slope of the line confirms a negative correlation between 'horsepower' and 'MPG', aligning with the expectation that cars with more horsepower generally have lower fuel efficiency. The fit of the line to the data indicates a suitable linear model for this relationship.
