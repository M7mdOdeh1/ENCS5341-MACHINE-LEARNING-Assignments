## 1. Model Selection and Hyper-parameters Tunning

```python
# Import necessary libraries for this section
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.linear_model import Ridge
```

### Part 1: Data Split and Visualization
In this code, I loaded the `data_reg.csv` dataset and splitted it into Training, Validation, and Testing sets with 120, 40, 40 rows respectively. Then, i plot the examples from the three sets in a scatter plot (3d plot):
   - The plot visualizes the distribution of the three subsets.
   - Axes represent `x1` (x-axis), `x2` (y-axis), and `y` (z-axis).
   - Each set is marked with different colors: red (training), green (validation), blue (testing).

```python
# Load the data from the CSV file
file_path = 'data_reg.csv'
data = pd.read_csv(file_path)

# Split the dataset into training, validation, and testing sets
train, remain = train_test_split(data, train_size=120, shuffle=False)
validation, test = train_test_split(remain, train_size=40, shuffle=False)

# Plot the data
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for training data
ax.scatter(train['x1'], train['x2'], train['y'], color='r', label='Training set')

# Scatter plot for validation data
ax.scatter(validation['x1'], validation['x2'], validation['y'], color='g', label='Validation set')

# Scatter plot for testing data
ax.scatter(test['x1'], test['x2'], test['y'], color='b', label='Testing set')

# Adding labels for axises
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.legend()

# Show plot
plt.show()
```

- This approach is an important way to analyze and model the data.
- The visualization assists in understanding how the data is distributed across the different sets, which is important for assessing the model's ability to generalize.

### Part 2: Polynomial Regression Analysis and Model Evaluation

This section involves applying polynomial regression to the training dataset and evaluating the performance of each model using the validation set. The goal is to determine the best polynomial degree for modeling the dataset.  

For each degree from 1 to 10, the `PolynomialFeatures` class is used to expand the original features into a polynomial feature space. Then, a `LinearRegression` model is trained on the polynomial features for each degree. The trained model is then used to predict on the validation set. The mean squared error (MSE) between the predictions and the true values is calculated as the validation error.

The resulting plot is used to determine the polynomial degree that achieves the lowest validation error, indicating the best model fit.



```python
# Apply polynomial regression for each degree from 1 to 10
degrees = range(1, 11)
validation_errors = []
models = []

X_train, y_train = train[['x1', 'x2']], train['y']
X_validation, y_validation = validation[['x1', 'x2']], validation['y']

for degree in degrees:
    # Create polynomial features
    poly = PolynomialFeatures(degree)
    X_train_polynomial = poly.fit_transform(X_train)
    X_validation_polynomial = poly.transform(X_validation)

    # Train the model
    model = LinearRegression()
    model.fit(X_train_polynomial, y_train)
    models.append(model)

    # Predict on the validation set
    y_validation_pred = model.predict(X_validation_polynomial)

    # Calculate and store the validation error
    mse = mean_squared_error(y_validation, y_validation_pred)
    validation_errors.append(mse)

# Plot the validation error vs polynomial degree curve
plt.figure(figsize=(10, 6))
plt.plot(degrees, validation_errors, marker='o')
plt.title('Validation Error vs Polynomial Degree')
plt.xlabel('Polynomial Degree')
plt.ylabel('Mean Squared Error')
plt.xticks(degrees)
plt.grid(True)
plt.show()
```

#### The Result:
A quadratic model (degree 2) minimizes the validation error, striking the optimal balance between bias and variance. This suggests it captures the trend without overfitting or underfitting and ensuring reliable generalization to unseen data.  
In addition, we can notice an increase in the error at degree 7 indicates potential overfitting at higher complexities

#### Plot the surface of the learned functions
The following code visualizes polynomial regression models for degrees 1 to 10, plotting the predicted surfaces and training data on a 3D plot. For each degree, it generates polynomial features, fits a model to the training data, and predicts over a feature grid. The results are displayed on a large figure with individual subplots for each degree, highlighting the model's complexity and fit to the training data with appropriate titles and labels for ease of interpretation.


```python
# Adjusting the mesh grid to a lower resolution to manage computational intensity
x1_range = np.linspace(X_train['x1'].min(), X_train['x1'].max(), 50)
x2_range = np.linspace(X_train['x2'].min(), X_train['x2'].max(), 50)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)

fig = plt.figure(figsize=(20, 40))

for i, degree in enumerate(range(1, 11)):
    # Create polynomial features for the current degree and fit them to the training data
    polynomial = PolynomialFeatures(degree)
    X_train_polynomial = polynomial.fit_transform(X_train)
    model = models[i]
    
    # Create a DataFrame for the grid data with feature names
    grid_data = pd.DataFrame(np.hstack((x1_grid.reshape(-1, 1), x2_grid.reshape(-1, 1))), columns=['x1', 'x2'])
    
    # Transform the grid data using the fitted PolynomialFeatures
    X_grid_polynomial = polynomial.transform(grid_data)

    # Predict over the grid
    y_grid_pred = model.predict(X_grid_polynomial).reshape(x1_grid.shape)

    # Plotting
    ax = fig.add_subplot(5, 2, i + 1, projection='3d')
    ax.plot_surface(x1_grid, x2_grid, y_grid_pred, alpha=0.5)
    ax.scatter(train['x1'], train['x2'], train['y'], color='r', marker='o', label='Training set')
    ax.set_title(f'Polynomial Degree {degree}')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('Y')
    ax.legend()

plt.tight_layout()
plt.show()
```

### Part 3: Ridge Regression

This code applies Ridge regression with an 8th-degree polynomial feature expansion to the training data, assessing model performance across different regularization strengths.


```python
# Preparing the polynomial features for degree 8
polynomial = PolynomialFeatures(degree=8)
X_train_polynomial = polynomial.fit_transform(X_train)
X_validation_polynomial = polynomial.transform(X_validation)

# Regularization parameter options
alpha_values = [0.001, 0.005, 0.01, 0.1, 10]
validation_errors_ridge = []

# Apply Ridge regression for different values of the regularization parameter
for alpha in alpha_values:
    ridge_model = Ridge(alpha=alpha)
    ridge_model.fit(X_train_polynomial, y_train)
    
    # Predict on the validation set and calculate the validation error
    y_validation_pred = ridge_model.predict(X_validation_polynomial)
    mse = mean_squared_error(y_validation, y_validation_pred)
    validation_errors_ridge.append(mse)

# Plotting the MSE vs regularization parameter
plt.figure(figsize=(10, 6))
plt.plot(alpha_values, validation_errors_ridge, marker='o')
plt.title('Validation MSE vs Regularization Parameter (Ridge Regression)')
plt.xlabel('Regularization Parameter (alpha)')
plt.ylabel('Mean Squared Error')
plt.xscale('log')  # Use logarithmic scale for better visualization
plt.grid(True)
plt.show()
```

the best regularization parameter (alpha) among the options for a polynomial of degree 8 appears to be `0.01`. This value corresponds to the lowest point on the curve, indicating the smallest MSE on the validation set and suggesting an optimal balance between the model's complexity and its performance, and it's the most effective for preventing overfitting while maintaining model accuracy, and balancing bias and variance in the model.

## 2. Logistic Regression

```python
# import neccessary libraries for this section
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
```

### Part 1: Logistic Regression model with a linear decision boundary
In this code, a logistic regression model is trained for binary classification using a dataset with features 'x1' and 'x2' and binary class labels 'C1' and 'C2'. The class labels are first converted to numerical form (0 and 1), and then the model is trained and evaluated for accuracy on both training and testing data. Finally, the decision boundary of the model is visualized on a plot, showing how the model distinguishes between the two classes based on the features, so, providing a visual representation of the model's effectiveness in classifying the data.


```python
# Load the data
train_data = pd.read_csv('train_cls.csv')
test_data = pd.read_csv('test_cls.csv')

# Convert class labels from 'C1', 'C2' to 0, 1
train_data['class'] = train_data['class'].map({'C1': 0, 'C2': 1})
test_data['class'] = test_data['class'].map({'C1': 0, 'C2': 1})

# Split the data into features and labels
X_train = train_data[['x1', 'x2']]
y_train = train_data['class']
X_test = test_data[['x1', 'x2']]
y_test = test_data['class']

# Train the logistic regression model
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Compute and print the accuracy on the training and test sets
train_accuracy = log_reg.score(X_train, y_train)
test_accuracy = log_reg.score(X_test, y_test)
print('=======================================')
print(f'Training Accuracy: {train_accuracy*100:.2f}%')
print(f'Test Accuracy: {test_accuracy*100:.2f}%')
print('=======================================')


# Create a mesh grid for plotting the decision boundary
x_min, x_max = X_train['x1'].min() - 1, X_train['x1'].max() + 1
y_min, y_max = X_train['x2'].min() - 1, X_train['x2'].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))

# Create a DataFrame from mesh grid for prediction
mesh_df = pd.DataFrame(np.c_[xx.ravel(), yy.ravel()], columns=['x1', 'x2'])

# Predict on the mesh grid DataFrame
Z = log_reg.predict(mesh_df)
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap='RdYlGn')
plt.scatter(X_train['x1'], X_train['x2'], c=y_train, edgecolors='k', cmap='RdYlGn')
plt.title('Logistic Regression Decision Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

The logistic regression model, with a training accuracy of `~66.1%` and testing accuracy of `~68.2%`, suggests potential underfitting. The linear decision boundary may be too simplistic for the dataset, implying the need for a more complex model or enhanced features.


### Part 2: Logistic Regression model with a quadratic decision boundary
This code trains a logistic regression model with a quadratic decision boundary on a binary classification dataset, converting `C1` and `C2` labels to numeric values. Polynomial features of degree 2 are used to enable non-linear classification. The model's effectiveness is assessed through training and testing accuracies, and the decision boundary is visualized in a customized color scheme to describe class differentiation.



```python
# Create polynomial features of degree 2 and transform training and test data
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Convert the transformed features into DataFrames
X_train_poly_df = pd.DataFrame(X_train_poly, columns=poly.get_feature_names_out(X_train.columns))
X_test_poly_df = pd.DataFrame(X_test_poly, columns=poly.get_feature_names_out(X_test.columns))

# Train the logistic regression model on the polynomial features DataFrame
log_reg_poly = LogisticRegression()
log_reg_poly.fit(X_train_poly_df, y_train)

# Compute the accuracy on the polynomial feature sets DataFrames
train_accuracy_poly = log_reg_poly.score(X_train_poly_df, y_train)
test_accuracy_poly = log_reg_poly.score(X_test_poly_df, y_test)

print('============================================')
print(f'Training Accuracy: {train_accuracy_poly*100:.2f}%')
print(f'Test Accuracy: {test_accuracy_poly*100:.2f}%')
print('============================================')

# Create a mesh grid for plotting the decision boundary
x1_min, x1_max = X_train['x1'].min() - 1, X_train['x1'].max() + 1
x2_min, x2_max = X_train['x2'].min() - 1, X_train['x2'].max() + 1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.02), np.arange(x2_min, x2_max, 0.02))

# Create a DataFrame from mesh grid for prediction
mesh_df = pd.DataFrame(np.c_[xx1.ravel(), xx2.ravel()], columns=['x1', 'x2'])
mesh_poly_df = pd.DataFrame(poly.transform(mesh_df), columns=poly.get_feature_names_out(['x1', 'x2']))

# Predict on the mesh grid DataFrame to obtain the decision boundary
Z_poly = log_reg_poly.predict(mesh_poly_df)
Z_poly = Z_poly.reshape(xx1.shape)

# Plot the quadratic decision boundary
plt.figure(figsize=(10, 6))
plt.contourf(xx1, xx2, Z_poly, alpha=0.8, cmap='RdYlGn')
plt.scatter(X_train['x1'], X_train['x2'], c=y_train, edgecolors='k', cmap='RdYlGn')
plt.title('Logistic Regression Quadratic Decision Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

The visualized output showcases a logistic regression model's quadratic decision boundary, achieving high training (96.8%) and testing (95.5%) accuracies, effectively separating the binary classes, indicating a good fit and strong generalization to unseen data.
